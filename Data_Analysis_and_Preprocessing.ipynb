## ***Convert "time" to HH:MM:SS formate***

### ***Convert "time" to HH:MM:SS formate for sheet2***

# ***Implement Custom Health Condition Guidelines for dishwashing machines***



*   **Good:** if (Temperature is within the range of 15 to 65 degrees and Vibration falls between 250 and 7500)
*   **Bad:** if (Temperature is below 10 or above 65) or (Vibration is below 150 or above 7900)
*   **Moderate:** otherwise categorized as Moderate.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define temperature thresholds
Low_temp_threshold = 10
High_temp_threshold = 70

# Define vibration thresholds
Low_vibration_threshold = 150
High_vibration_threshold = 7900
filtered_data = Main_data[Main_data['Vibration Sensor-1'] != 2147483647.000]
# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Temperature Sensor-1', y='Vibration Sensor-1', data=filtered_data, alpha=0.7, label='Good health')

# apply Conditions
moderate_Health = filtered_data[
    ((10 < filtered_data['Temperature Sensor-1']) & (filtered_data['Temperature Sensor-1'] <= 15)) |
    ((65 <= filtered_data['Temperature Sensor-1']) & (filtered_data['Temperature Sensor-1'] <= 70)) |
    (7500 < filtered_data['Vibration Sensor-1']) & (filtered_data['Vibration Sensor-1'] < 7900) |
    (150 < filtered_data['Vibration Sensor-1']) & (filtered_data['Vibration Sensor-1'] <= 300)
]


# Plot moderate temperature points
sns.scatterplot(x='Temperature Sensor-1', y='Vibration Sensor-1', data=moderate_Health, color='orange', label='Moderate Health')


# Filter data for bad health
bad_health_data = filtered_data[
    (filtered_data['Temperature Sensor-1'] > High_temp_threshold) |
    (filtered_data['Temperature Sensor-1'] < Low_temp_threshold) |
    (filtered_data['Vibration Sensor-1'] > High_vibration_threshold) |
    (filtered_data['Vibration Sensor-1'] < Low_vibration_threshold)
]


# Plot bad health points
sns.scatterplot(x='Temperature Sensor-1', y='Vibration Sensor-1', data=bad_health_data, color='red', label='Bad Health')


plt.title('Vibration Sensor-1 vs Temperature')
plt.xlabel('Temperature Sensor-1')
plt.ylabel('Vibration Sensor-1')
plt.legend()
plt.show()
```

# ***Implement Custom Health Condition Guidelines for dishwashing machines***



*   **Good:** if (Temperature is within the range of 15 to 65 degrees and Vibration falls between 250 and 7500)
*   **Bad:** if (Temperature is below 10 or above 65) or (Vibration is below 150 or above 7900)
*   **Moderate:** otherwise categorized as Moderate.

```python
Main_data['Health'] = 'Good'  # Initialize all rows with 'Good'
```

```python
# Conditions for 'Bad'
bad_condition = ((Main_data['Temperature Sensor-1'] > 70) | (Main_data['Temperature Sensor-1'] < 10) |
                 (Main_data['Vibration Sensor-1'] > 7900) | (Main_data['Vibration Sensor-1'] < 150) |
                 (Main_data['Vibration Sensor-2'] > 7900) | (Main_data['Vibration Sensor-2'] < 150) |
                 (Main_data['Vibration Sensor-3'] > 7900) | (Main_data['Vibration Sensor-3'] < 150))

# Conditions for 'Moderate'
Moderate_condition = ((Main_data['Temperature Sensor-1'] > 65) &  (Main_data['Temperature Sensor-1'] <= 70) |
                  (Main_data['Temperature Sensor-1'] > 10) &  (Main_data['Temperature Sensor-1'] <= 15) |
                  (Main_data['Vibration Sensor-1'] > 7500) &  (Main_data['Vibration Sensor-1'] <= 7900) |
                  (Main_data['Vibration Sensor-1'] > 150) &  (Main_data['Vibration Sensor-1'] <= 250) |
                  (Main_data['Vibration Sensor-2'] > 7500)&  (Main_data['Vibration Sensor-2'] <= 7900) |
                  (Main_data['Vibration Sensor-2'] > 150) &  (Main_data['Vibration Sensor-2'] <= 300) |
                  (Main_data['Vibration Sensor-3'] > 7500) &  (Main_data['Vibration Sensor-3'] <= 7900) |
                  (Main_data['Vibration Sensor-3'] > 150) & (Main_data['Vibration Sensor-3'] <= 300))
```

```python
# Updating Health column based on conditions
Main_data.loc[bad_condition, 'Health'] = 'Bad'
Main_data.loc[Moderate_condition, 'Health'] = 'Moderate'

Main_data
```
